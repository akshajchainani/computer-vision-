###Aim: To detect and visualise the contours of objects within an image.

###Theory:
Contour detection is a key process in computer vision, where the goal is to identify and trace the boundaries of objects within an image. This method is essential for determining the shape, size, and structure of objects, making it valuable in various tasks like object detection, image segmentation, and shape analysis.

1 Contour Representation:
A contour is a curve that connects continuous points along a boundary with the same intensity or color in a 2D image. Contours can be represented as sequences of points or as polygonal chains, which define the shape of objects within an image and enable precise boundary detection.

2 Contour Detection Algorithms:
Canny Edge Detection:
The Canny Edge Detection algorithm is widely used for detecting edges, which often correspond to contours. The process involves:

Noise Reduction: Smoothing the image using a Gaussian filter to minimize noise.
Gradient Calculation: Computing the gradient intensity to identify areas with significant intensity changes, indicating edges.
Non-Maximum Suppression: Eliminating non-essential edge responses, leaving only the strongest edge pixels, resulting in thin, sharp edges.
Edge Tracking by Hysteresis: Finalizing edges by eliminating weak edges not connected to strong ones.
Sobel Operator:
The Sobel operator is a convolutional filter used to calculate the gradient of image intensity, highlighting areas with high spatial frequency, typically corresponding to edges. It applies two 3x3 kernels to the image to detect intensity changes in both the horizontal and vertical directions.

Finding Contours in Binary Images:
After detecting edges, contours can be extracted from a binary image, where edges are marked as 1, and non-edges as 0. The findContours function in OpenCV is commonly used for this purpose, retrieving contours based on edge information and representing them in different formats, such as points or hierarchical structures.

3 Contour Approximation:
Contour approximation involves simplifying a contour by reducing the number of points while maintaining the contour's overall shape. The Douglas-Peucker algorithm is commonly used for this purpose, approximating a contour as a polygon with fewer vertices, which can reduce computational complexity in further processing.

4 Hierarchical Contours:
Contours can sometimes be nested within one another, creating a hierarchical structure. For instance, a doughnut-shaped object has an outer contour representing its shape and an inner contour representing the hole. Understanding contour hierarchy is important for analyzing complex shapes or objects with internal features.

Applications:
Shape Analysis:
Contours are essential for analyzing the shape of objects, which is fundamental for object recognition, classification, and measurement. By studying contours, key shape descriptors such as area, perimeter, and moments can be derived.

Object Detection and Recognition:
Contour detection is crucial for determining the exact location, size, and shape of objects within an image. This information is critical for object recognition systems, which use these attributes to identify objects accurately.

Image Segmentation:
Contours often serve as the foundation for segmenting images into different regions. This allows for detailed analysis of specific parts of an image, such as isolating objects from the background for further processing or analysis.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread("/content/drive/MyDrive/colab for cvr/cont_new.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image)
```

```python
ret,thresh = cv2.threshold(image_gray,127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
```

```python
contour_img = cv2.drawContours(image, contours, -1, (0, 255, 0), 4)
plt.imshow(contour_img)
```
